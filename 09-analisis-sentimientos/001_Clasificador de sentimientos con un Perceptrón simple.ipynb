En esta actividad vas a implementar desde cero un modelo de Perceptr贸n en Python usando solo NumPy. Vamos a entrenarlo con frases breves escritas en espa帽ol rioplatense para que aprenda a reconocer si una frase tiene un sentimiento positivo o negativo.

El objetivo es entender c贸mo funciona una neurona artificial b谩sica, c贸mo se ajustan sus pesos durante el aprendizaje, y c贸mo puede hacer predicciones en base a un conjunto peque帽o de frases.

## 1. Datos de entrenamiento
Vamos a usar un peque帽o conjunto de frases etiquetadas como positivas (1) o negativas (0). Las frases son simples como las que podr铆amos encontrar en una conversaci贸n cotidiana.

Tambi茅n definimos un vocabulario b谩sico de palabras clave que aparecen con frecuencia y que nos pueden ayudar a inferir el sentimiento.

```python
import numpy as np

# Frases con su etiqueta de sentimiento (1 = positivo, 0 = negativo)
frases = [
    "Amo el verano en Buenos Aires",
    "No me gusta el tr谩fico matutino",
    "Este asado est谩 espectacular",
    "Qu茅 baj贸n, perd铆 el colectivo",
    "Me encanta salir los domingos",
    "Detesto el calor h煤medo"
]

etiquetas = np.array([1, 0, 1, 0, 1, 0])  # Etiquetas

# Vocabulario manual con palabras claves de carga emocional
vocabulario = ["amo", "no", "gusta", "asado", "espectacular", "baj贸n", "perd铆", "detesto", "calor", "h煤medo"]
```

##О 2. Representaci贸n num茅rica: Vectorizaci贸n de frases
Vamos a convertir cada frase en un vector binario que indica si una palabra del vocabulario aparece o no.

```python
# Funci贸n que convierte una frase en un vector binario seg煤n el vocabulario
def vectorizar(frase, vocabulario):
    tokens = frase.lower().split()
    return np.array([1 if palabra in tokens else 0 for palabra in vocabulario])

# Aplicamos la funci贸n a todas las frases
X = np.array([vectorizar(frase, vocabulario) for frase in frases])
```

## 3. Definici贸n del modelo: el Perceptr贸n
Un perceptr贸n es una funci贸n que:

* Multiplica cada entrada por un peso.

* Suma los resultados y agrega un sesgo (bias).

* Aplica una funci贸n de activaci贸n (en este caso, escal贸n) para decidir si "dispara" o no.

Vamos a inicializar los pesos aleatoriamente y entrenar el modelo para que aprenda de los errores.

```python
# Inicializamos pesos y bias
np.random.seed(42)
pesos = np.random.randn(len(vocabulario))
bias = 0.0

# Funci贸n de activaci贸n (escal贸n)
def activar(suma):
    return 1 if suma > 0 else 0

# Funci贸n de predicci贸n
def predecir(x):
    suma = np.dot(x, pesos) + bias
    return activar(suma)
```

## 4. Entrenamiento del modelo
Ahora vamos a entrenar el perceptr贸n ajustando los pesos seg煤n los errores que comete. Usamos la regla de aprendizaje del perceptr贸n: si se equivoca, ajusta los pesos y el sesgo para acercarse al resultado correcto.

```python
# Par谩metros de entrenamiento
tasa_aprendizaje = 0.1
epocas = 20

# Bucle de entrenamiento
for epoca in range(epocas):
    errores = 0
    for i in range(len(X)):
        x_i = X[i]
        y_real = etiquetas[i]
        y_pred = predecir(x_i)
        error = y_real - y_pred
        if error != 0:
            pesos += tasa_aprendizaje * error * x_i
            bias += tasa_aprendizaje * error
            errores += 1
    print(f"poca {epoca + 1}: Errores = {errores}")
```

##И 5. Prueba con nuevas frases
Ahora vamos a ver c贸mo se comporta nuestro perceptr贸n con frases nuevas que no vio durante el entrenamiento.

```python
# Frases nuevas para testeo
frases_prueba = [
    "No aguanto este calor",
    "Qu茅 hermoso d铆a para pasear",
    "Detesto levantarme temprano"
]

# Vectorizamos las frases nuevas
X_prueba = np.array([vectorizar(frase, vocabulario) for frase in frases_prueba])
predicciones = [predecir(x) for x in X_prueba]

# Mostramos los resultados
for frase, pred in zip(frases_prueba, predicciones):
    resultado = "Positivo" if pred == 1 else "Negativo"
    print(f"Frase: '{frase}' => Sentimiento predicho: {resultado}")
```
