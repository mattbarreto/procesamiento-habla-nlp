```python
# Instalar transformers y accelerate si no están
!pip install -q transformers accelerate

from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline

# Cargar modelo y tokenizer
model_id = "microsoft/Phi-3-mini-4k-instruct"

model = AutoModelForCausalLM.from_pretrained(
    model_id,
    device_map="auto",  # Se adapta automáticamente al entorno (Colab suele tener GPU)
    torch_dtype="auto",
    trust_remote_code=True,
)

tokenizer = AutoTokenizer.from_pretrained(model_id)

# Crear pipeline para generación de texto
generator = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    return_full_text=False,
    max_new_tokens=500,
    do_sample=False
)

# Instrucción
prompt = "Contá un chiste gracioso sobre gallinas."

# Los modelos tipo instruct aceptan prompts como texto plano.
output = generator(prompt)

# Mostrar el resultado
print(output[0]["generated_text"])
```
